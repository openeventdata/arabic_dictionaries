```python

#generating the entry
def gen_person_entry(wiki_entry): 
    
    
    #Entry now is dictionary {"en_title" : "" , "en_pageid" : ""  , "ar_title" : "" , "ar_pageid" : ""}
    has_arabic_data = True if wiki_entry['ar_title'] != '' else False 
    has_english_data = True if wiki_entry['en_title'] != '' else False 
    
    #special cases that might occur because of external errors in scrapping actors names
    if has_arabic_data : 
        if 'List of' in wiki_entry['ar_title']:
            print(wiki_entry , "---> DISCARDED")
            return -1 
    if has_english_data : 
        if 'List of' in wiki_entry['en_title'] :
            print(wiki_entry , "---> DISCARDED")
            return -1 
        
    entry ={}        
    
    #make a request to get wiki_links : en,es , ar (en and es are the same)
    entry['names'] = []
    names = {} 
    names['english'] = wiki_entry['en_title'] if has_english_data else ''
    names['arabic']= wiki_entry['ar_title'] if has_arabic_data else ''
    
    entry['names'].append(names)
    
    
    entry['wiki_scrape_date'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
#     entry['instance_of'] = 'person'
    
    #page id
    entry['pageid'] =[]
    pages_id = {} 
    pages_id['english'] = wiki_entry['en_pageid'] if has_english_data else ''
    pages_id['arabic'] = wiki_entry['ar_pageid'] if has_arabic_data else ''
    
    
    entry['pageid'].append(pages_id)

    
    
#     #usings the wiki api, the name is the same as wikipedia link 
#     entry['wikipedia'] = {}
#     entry['wikipedia']['english'] = person_name
#     entry['wikipedia']['spanish'] = person_name
#     entry['wikipedia']['arabic']= person_wiki_links[0]['*'] if has_arabic_data else []
    
    #TODO : havested from , wiki_categories
    entry['harvested_from'] = []
    havested_from= {} 
    
    havested_from['english'] = "https://en.wikipedia.org/wiki/" +wiki_entry['en_title'].replace(' ', '_') if has_english_data else ''
    havested_from['arabic'] = "https://ar.wikipedia.org/wiki/" +wiki_entry['ar_title'].replace(' ', '_') if has_arabic_data else ''
    entry['harvested_from'].append(havested_from)
    
    #get wiki roles info
    entry['wiki_roles'] =[]
    #.. english 
    wiki_roles ={}
    wiki_roles['english'] = get_en_wiki_roles2(wiki_entry['en_title']) if has_english_data else []
    
    #.. arabic
    wiki_roles['arabic'] = get_ar_wiki_roles2(wiki_entry['ar_title']) if has_arabic_data else []
    entry['wiki_roles'].append(wiki_roles)
    
    if wiki_roles['english'] == [] and wiki_roles['arabic'] ==[] : #discard this 
        print(wiki_entry , "---> DISCARDED")
        return -1 
    

    #cameo coding : TODO
    entry['cameo_coding'] = []
    print(wiki_entry , "---> OK!")
    #return the entry
    return entry

```

```python
#read data from file 
# every data_entry is a dictioin
#process 

import json
# from tqdm import tqdm 

generated_entries = [] # entries that have useful role information 
discard_entries  = [] # entries that don't have role info 
    
with open('wiki_actors_3', 'r') as f:
        actors = json.load(f)

for actor in actors : 
    process = gen_person_entry(actor)
    if process == -1 : 
        discard_entries.append(actor)
    else : 
        generated_entries.append(process)
```
