```python
import spacy 
import re
import nltk
import pyarabic as araby
text = 'ومثلما يبدو الاحتفال بعيدين في ذات اليوم أمراً فريدا، فإن السجالات التي تدور حول "ثورة 25 يناير" وما تبقى منها، وما تحقق من أهدافها، لا تخلو من الخصوصية والاستثناء. فالبعض يراها تتقدم، وتحقق أهدافها، والبعض الآخر يراها مختطفة، ومجهضة، وتأتي الآراء غالبا، حسب موقف صاحبا من السلطة الحاكمة حاليا. وما بين هذا وذاك، تتجلى مشاهد عالقة في الأذهان، وموثقة بالصورة ومقاطع الفيديو، لا يمكن إنكارها، أو نفي أنها أصدق ما تبقى من هذه الحدث السياسي والاجتماعي الفريد من نوعه. '

print(text)
```

## spacy

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
for tok in doc : 
    print(tok.text)
```

## NLTK 

```python
wordsArray = nltk.word_tokenize(text)
#wordsArray = nltk.wordpunct_tokenize(lex)
```

```python
wordsArray
```

## white-space tokenizer

```python
tokens = text.split(' ')
tokens
```

## PyArabic

```python
from pyarabic.araby import tokenize, is_arabicrange, strip_tashkeel
tokens = tokenize(text)
tokens
```

## using a regex 

```python
tokens = re.findall(u'[\u0600-\u06FF0-9]+', text)
tokens
```
